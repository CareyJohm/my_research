```python
nor_ha_sub_list = []

for i in range(len(epic_list)):
    for a in range(len(df2['HA663-HA656'])):
        if epic_list[i] == df2['EPIC'][a]:
            nor_ha_sub_list.append(abs(df2['HA663-HA656'][a]/df2['HA663'][a]))
```
